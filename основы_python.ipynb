#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(text):
    vowels = "aeiou"
    count = 0
    text_lower = text.lower()
    
    for char in text_lower:
        if char in vowels:
            count += 1
            
    return count


def test_count_vowels():
    assert count_vowels("Hello World") == 3
    assert count_vowels("aeiou") == 5
    assert count_vowels("bcdfg") == 0
    assert count_vowels("") == 0
    assert count_vowels("AEIOU") == 5
    
    print("Все тесты пройдены успешно!")


if __name__ == "__main__":
    test_count_vowels()
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def has_unique_chars(string):
    return len(string) == len(set(string))

def test_has_unique_chars():
    assert has_unique_chars("abcde") == True
    assert has_unique_chars("hello") == False
    assert has_unique_chars("") == True
    assert has_unique_chars("a") == True
    assert has_unique_chars("abcda") == False
    
    print("Все тесты пройдены успешно!")

if __name__ == "__main__":
    test_has_unique_chars()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_bits(n):
    return bin(n).count('1')

def test_count_bits():
    assert count_bits(0) == 0
    assert count_bits(1) == 1
    assert count_bits(2) == 1
    assert count_bits(3) == 2
    assert count_bits(7) == 3
    
    print("Все тесты пройдены успешно!")

if __name__ == "__main__":
    test_count_bits()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    count = 0
    while n >= 10:
        product = 1
        for digit in str(n):
            product *= int(digit)
        n = product
        count += 1
    return count

def test_magic():
    assert magic(39) == 3
    assert magic(4) == 0
    assert magic(999) == 4
    assert magic(25) == 2
    assert magic(10) == 1
    
    print("Все тесты пройдены успешно!")

if __name__ == "__main__":
    test_magic()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    n = len(pred)
    total = 0
    for i in range(n):
        total += (pred[i] - true[i]) ** 2
    return total / n

def test_mse():
    assert mse([1, 2, 3], [1, 2, 3]) == 0
    assert mse([1, 2, 3], [2, 3, 4]) == 1
    assert mse([0, 0, 0], [1, 1, 1]) == 1
    assert mse([1, 2], [3, 4]) == 4
    assert mse([5, 6, 7], [5, 6, 8]) == 1/3
    
    print("Все тесты пройдены успешно!")

if __name__ == "__main__":
    test_mse()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factors(n):
    factors = {}
    divisor = 2
    while n > 1:
        while n % divisor == 0:
            factors[divisor] = factors.get(divisor, 0) + 1
            n //= divisor
        divisor += 1
    
    result = ""
    for prime in sorted(factors.keys()):
        power = factors[prime]
        if power == 1:
            result += f"({prime})"
        else:
            result += f"({prime}**{power})"
    
    return result

def test_prime_factors():
    assert prime_factors(86240) == "(2**5)(5)(7**2)(11)"
    assert prime_factors(2) == "(2)"
    assert prime_factors(12) == "(2**2)(3)"
    assert prime_factors(13) == "(13)"
    assert prime_factors(100) == "(2**2)(5**2)"
    
    print("Все тесты пройдены успешно!")

if __name__ == "__main__":
    test_prime_factors()
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    total = 0
    k = 1
    while total < number:
        total += k * k
        if total == number:
            return k
        k += 1
    return "It is impossible"

def test_pyramid():
    assert pyramid(1) == 1
    assert pyramid(5) == 2
    assert pyramid(14) == 3
    assert pyramid(30) == 4
    assert pyramid(55) == 5
    assert pyramid(3) == "It is impossible"
    assert pyramid(10) == "It is impossible"
    assert pyramid(25) == "It is impossible"
    
    print("Все тесты пройдены успешно!")

if __name__ == "__main__":
    test_pyramid()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(number):
    digits = str(number)
    n = len(digits)
    
    if n % 2 == 0:
        mid = n // 2
        left_sum = sum(int(d) for d in digits[:mid-1])
        right_sum = sum(int(d) for d in digits[mid+1:])
    else:
        mid = n // 2
        left_sum = sum(int(d) for d in digits[:mid])
        right_sum = sum(int(d) for d in digits[mid+1:])
    
    return left_sum == right_sum

def test_is_balanced():
    assert is_balanced(12321) == True
    assert is_balanced(123321) == True
    assert is_balanced(12345) == False
    assert is_balanced(1234) == False
    assert is_balanced(1) == True
    assert is_balanced(11) == True
    assert is_balanced(121) == True
    
    print("Все тесты пройдены успешно!")

if __name__ == "__main__":
    test_is_balanced()
```
